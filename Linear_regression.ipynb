```python
#manipulação de dados
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 

#REGRESSÃO LINEAR
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import cross_val_score

X_test = np.load('X_test_regression1.npy')
X_test_df = pd.DataFrame(X_test)
X_train = np.load('X_train_regression1.npy')
X_train_df = pd.DataFrame(X_train)
Y_train = np.load('Y_train_regression1.npy')
Y_train_df = pd.DataFrame(Y_train)
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

#LINEAR REGRESSION
regLinear = linear_model.LinearRegression()
regLinear.fit(X_train, Y_train)
pred_regLinear = regLinear.predict(X_train)
print("Coefficients: \n",regLinear.coef_)
```

```python
print("Mean squared error: %.2f" % mean_squared_error(Y_train, pred_regLinear ))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(Y_train, pred_regLinear ))
```
